```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Carprices.xlsx')
df
```

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
final = merged.drop(['Car Model','Mercedez Benz C class'], axis = 'columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final.drop('Sell Price($)', axis='columns')
X
```

```python
y = final['Sell Price($)']
y
```

```python
model.fit(X,y)
```

```python
# Price of mercedez benz that is 4 yr old with mileage 45000
model.predict([[45000,4,0,0]])
```

```python
#Predict price of a BMW X5 that is 7 yr old with mileage 86000
model.predict([[86000,7,0,1]])
```

```python
#m
model.coef_
```

```python
#b
model.intercept_
```

# First plot data points on a scatter plot chart to see if linear regression model can be applied

```python
plt.xlabel("Mileage", fontsize=20)
plt.ylabel("Price", fontsize=20)
plt.title("Car Price Analysis", fontsize=20)

plt.scatter(df['Car Model'], final['Sell Price($)'], color='red', marker='+')
plt.plot(df['Car Model'], model.predict(final[['Mileage','Age(yrs)','Audi A5','BMW X5']]), color='blue')
```
