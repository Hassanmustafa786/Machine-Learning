```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color='green',linewidth=5,linestyle='dashdot',alpha=0.5)
```

```python
plt.plot(x,y, "rD--")
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [50,43,46,13,75,97,24]
avg_t = [32,53,67,39,54,86,41]
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title('Weather')

plt.plot(days,max_t, label= 'Max')
plt.plot(days,min_t, label= 'Min')
plt.plot(days,avg_t, label= 'Avg')

plt.legend(loc='best', shadow=True, fontsize='small')
plt.grid()
```

```python
import numpy as np
```

```python
comp =['Google','Amazon','Microsoft','Facebook']
rev = [93,62,43,85]
profit = [60,30,3,52]
```

```python
plt.title('US Tech Stocks')
plt.ylabel('Revenue (bln)')
plt.xlabel('Company')

plt.bar(comp,rev, label= 'Revenue')
plt.bar(comp,profit, label= 'Profit')
plt.legend()
```

```python
x_pos = np.arange(len(comp))
x_pos
```

```python
plt.xticks(x_pos,comp)
plt.title('US Tech Stocks')
plt.ylabel('Revenue (bln)')
plt.xlabel('Company')

plt.bar(x_pos-0.2, rev, width=0.4, label='Revenue')
plt.bar(x_pos+0.2, profit, width=0.4, label='Profit')
plt.legend()
```

```python
plt.yticks(x_pos,comp)
plt.title('US Tech Stocks')

plt.barh(x_pos-0.2, rev, label='Revenue')
plt.barh(x_pos+0.2, profit, label='Profit')
plt.legend()
```

```python
#For histogram you need only single dimensional array
blood_sugar = [50,13,27,39,90,64,78,41,94,58,35,65,24]

plt.hist(blood_sugar, bins=[40,50,60,70], rwidth=0.95, color='g',histtype='step')
```

```python
blood_sugar_men = [50,13,27,39,90,64,78,41,94,58,35,65,24]
blood_sugar_women = [40,39,65,80,51,83,79,29,43,61,37,10,5]

plt.hist([blood_sugar_men, blood_sugar_women], bins=[40,50,60,70], rwidth=0.95, color=['green','orange'],
         label=['Men','Women'])

plt.xlabel('Sugar')
plt.ylabel('No. of Patient')
plt.title('Monthly Sugar Analysis')
plt.legend(loc='best')
```
